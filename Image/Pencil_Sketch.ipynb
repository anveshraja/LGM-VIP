# Task:
We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.

```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
```

```python
#loading image
img=cv2.imread('Kane.jpg',1)
img=cv2.resize(img,(400,400))
display(Image.fromarray(img))
```

```python
img1=cv2.imread('Kane.jpg')
img1=cv2.resize(img1,(400,400))
#cv2.imshow("KANE",img1)
#cv2.waitKey(0)
display(Image.fromarray(img1))
```

```python
#converting the image into gray scale
img1_gray=cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img1_gray=cv2.resize(img1_gray,(400,400))
#cv2.imshow("Gray_Kane",img1_gray)
#cv2.waitKey(0)
display(Image.fromarray(img1_gray))
```

```python
#inverting the image to enhace details
img1_inv = 255 - img1_gray
#cv2.imshow("Inverted", img1_inv)
#cv2.waitKey(0)
display(Image.fromarray(img1_inv))
```

```python
#creating blurry image of the gray scale image
img1_blurred = cv2.GaussianBlur(img1_inv, (21, 21), 0)
#cv2.imshow("Inverted", img1_blurred)
#cv2.waitKey(0)
display(Image.fromarray(img1_blurred))
```

```python
#converting blurry image into inverted image and dividing with gray image to get a pencil sketch
inv_blurred=255-img1_blurred
img1_pencil=cv2.divide(img1_gray,inv_blurred,scale=256.0)
#cv2.imshow("Pencil Sketch",img1_pencil)
#cv2.waitKey(0)
display(Image.fromarray(img1_pencil))
```

```python
#viewing original and pencil sketch images
#cv2.imshow("Original image", img1)
#cv2.imshow("Pencil sketch", img1_pencil)
#cv2.waitKey(0)
display(Image.fromarray(img1))
display(Image.fromarray(img1_pencil))
```
